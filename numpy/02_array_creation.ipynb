```python
import numpy as np
```

## NumPy Array Creation Functions

### 1. `np.linspace()`: Create an array with evenly spaced numbers over a specified interval.

```python
# Create 5 evenly spaced numbers from 0 to 10
linspace_array = np.linspace(0, 10, 5)
print("linspace_array:", linspace_array)
```

### 2. `np.logspace()`: Create an array with numbers that are evenly spaced on a log scale.

```python
# Create 5 evenly spaced numbers on a log scale from 10^0 to 10^2
logspace_array = np.logspace(0, 2, 5)
print("logspace_array:", logspace_array)
```

### 3. `np.full()`: Create an array of a given shape and fill it with a specified value.

```python
# Create a 2x3 array filled with 7s
full_array = np.full((2, 3), 7)
print("full_array:\n", full_array)
```

### 4. `np.eye()`: Create a 2-D array with ones on the diagonal and zeros elsewhere (identity matrix).

```python
# Create a 3x3 identity matrix
eye_array = np.eye(3)
print("eye_array:\n", eye_array)
```

### 5. `np.identity()`: Create a square identity matrix.

```python
# Create a 4x4 identity matrix
identity_array = np.identity(4)
print("identity_array:\n", identity_array)
```

## Random Arrays

### 1. Vectors (1D Arrays)

```python
# Create a 1D array (vector) of 5 random numbers between 0 and 1
random_vector = np.random.rand(5)
print("random_vector:", random_vector)
```

### 2. Matrices (2D Arrays)

```python
# Create a 2x3 matrix of random numbers
random_matrix = np.random.rand(2, 3)
print("random_matrix:\n", random_matrix)
```

### 3. 3D Arrays

```python
# Create a 2x2x2 3D array of random integers between 0 and 9
random_3d_array = np.random.randint(0, 10, size=(2, 2, 2))
print("random_3d_array:\n", random_3d_array)
```

## NumPy Array Attributes

```python
# Let's use this array for demonstrating attributes
example_array = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
print("Example Array:\n", example_array)
```

### 1. `shape`: The dimensions of the array as a tuple.

```python
print("Shape of example_array:", example_array.shape)
```

### 2. `ndim`: The number of dimensions of the array.

```python
print("Number of dimensions (ndim) of example_array:", example_array.ndim)
```

### 3. `size`: The total number of elements in the array.

```python
print("Total number of elements (size) in example_array:", example_array.size)
```

### 4. `dtype`: The data type of the elements in the array.

```python
print("Data type (dtype) of elements in example_array:", example_array.dtype)
```

### 5. `itemsize`: The size in bytes of each element of the array.

```python
print("Size in bytes of each element (itemsize) in example_array:", example_array.itemsize)
```

### 6. `nbytes`: The total number of bytes consumed by the array data.

```python
print("Total bytes consumed by array data (nbytes) in example_array:", example_array.nbytes)
```
