# NumPy Basics: Array Creation
First, we need to import the NumPy library. It is common practice to import it as `np`.

```python
import numpy as np
```

### 1. np.array()
This is used to create an array from a Python list.

```python
# Creating a simple 1D array
my_list = [1, 2, 3, 4, 5]
arr = np.array(my_list)
print("np.array example:", arr)
```

### 2. np.arange()
This creates an array with a range of numbers, similar to Python's built-in `range()`.

```python
# Start at 0, end before 10
range_arr = np.arange(10)
print("np.arange example:", range_arr)

# Start at 2, end before 10, step by 2
step_arr = np.arange(2, 10, 2)
print("np.arange with step:", step_arr)
```

### 3. np.zeros() and np.ones()
These are useful for creating placeholder arrays filled with either 0s or 1s.

```python
# Array of five zeros
zeros_arr = np.zeros(5)
print("np.zeros:", zeros_arr)

# A 2x3 matrix of ones
ones_arr = np.ones((2, 3))
print("np.ones (2x3 matrix):\n", ones_arr)
```
