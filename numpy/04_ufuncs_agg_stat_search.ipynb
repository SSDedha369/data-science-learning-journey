```python
import numpy as np

data = np.array([1, 4, 9, -16, 25])
print("Original:", data)

# Square root (handling only positive)
print("Sqrt:", np.sqrt(np.abs(data)))

# Logarithm
print("Log:", np.log(np.array([1, np.e, np.e**2])))

# Exponential
print("Exp:", np.exp(np.array([0, 1, 2])))

# Trigonometric
angles = np.array([0, np.pi/4, np.pi/2])
print("Sin:", np.sin(angles))
print("Cos:", np.cos(angles))

# Absolute value
print("Abs:", np.abs(data))

# Multi-dimensional example
matrix = np.array([[-1, -2], [3, 4]])
print("Matrix Abs:\n", np.abs(matrix))
print("Matrix Exp:\n", np.exp(matrix))
```

```python


arr = np.array([10, 20, 30, 40, 50])
print(f"Sum: {np.sum(arr)}")
print(f"Mean: {np.mean(arr)}")
print(f"Median: {np.median(arr)}")
print(f"Min: {np.min(arr)}")
print(f"Max: {np.max(arr)}")
print(f"Std Dev: {np.std(arr):.2f}")
print(f"Variance: {np.var(arr):.2f}")

# Percentiles and Quantiles
print(f"75th Percentile: {np.percentile(arr, 75)}")
print(f"0.5 Quantile: {np.quantile(arr, 0.5)}")

# Correlation and Covariance
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])
print("Correlation Matrix:\n", np.corrcoef(x, y))
print("Covariance Matrix:\n", np.cov(x, y))
```

```python


arr = np.array([10, 50, 20, 50, 30])

# np.where
print("Indices where > 25:", np.where(arr > 25))
print("Value replacement:", np.where(arr > 25, "High", "Low"))

# np.argmax and np.argmin
print("Index of max:", np.argmax(arr))
print("Index of min:", np.argmin(arr))

# np.nonzero
sparse_arr = np.array([1, 0, 2, 0, 3])
print("Indices of non-zero elements:", np.nonzero(sparse_arr))

# 2D Example
matrix = np.array([[0, 1], [1, 0]])
print("2D Nonzero indices:\n", np.nonzero(matrix))
```
