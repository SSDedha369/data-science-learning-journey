```python
import numpy as np

# 1D Array Iteration
arr1 = np.array([1, 2, 3])
for x in arr1:
    print(x)
```

```python
# 2D Array Iteration
arr2 = np.array([[1, 2, 3], [4, 5, 6]])
for row in arr2:
    for x in row:
        print(x)
```

```python
# Using nditer()
for x in np.nditer(arr2):
    print(x)
```

```python
# View: shared data
arr_view = arr1.view()
arr_view[0] = 100
print("Original:", arr1)
print("View:", arr_view)
```

```python
# Copy: independent data
arr_copy = arr1.copy()
arr_copy[0] = 999
print("Original:", arr1)
print("Copy:", arr_copy)
```

```python
# Dot product
a = np.array([1, 2])
b = np.array([3, 4])
print("Dot:", np.dot(a, b))
```

```python
# Matrix multiplication
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
print("Matmul:", A @ B)
```

```python
# Determinant
det = np.linalg.det(A)
print("Determinant:", det)
```

```python
# Inverse
inv = np.linalg.inv(A)
print("Inverse:", inv)
```

```python
# Eigenvalues and Eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)
print("Eigenvalues:", eigenvalues)
```
